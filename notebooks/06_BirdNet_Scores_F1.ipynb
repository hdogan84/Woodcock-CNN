```python
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
import numpy as np

dataset1_score = np.load('dataset1_f1.npz')
dataset2_score = np.load('holderried_f1.npz')
```

```python
thr1 = dataset1_score["thr"]
f1_ds1 = dataset1_score["f1"]

thr2 = dataset2_score["thr"]
f1_ds2 = dataset2_score["f1"]
```

```python
import matplotlib.pyplot as plt

# Create figure
plt.figure(figsize=(6, 4))

# Plot curves
plt.plot(thr1, f1_ds1, color='blue', linestyle='-',
    linewidth=2,
    label='Dataset 1'
)

plt.plot(
    thr2,
    f1_ds2,
    color='red',
    linestyle='--',
    linewidth=2,
    label='Dataset 2'
)

# Labels
plt.xlabel('Confidence threshold')
plt.ylabel(r'$\mathrm{F}_1$ score')

# Optional title (often omitted in papers because the caption serves as title)
# plt.title('Comparison of Two Datasets')

# Axis limits
plt.xlim([0, 1])
plt.ylim([0, 1.01])

# Grid and legend
plt.grid(True)
plt.legend()

# Improve layout
plt.tight_layout()

# Save in multiple formats
plt.savefig(
    "f1_comparison.png",
    dpi=300,
    bbox_inches="tight"
)

plt.savefig(
    "f1_comparison.tiff",
    dpi=600,
    bbox_inches="tight"
)

plt.savefig("f1_comparison.pdf", bbox_inches="tight")
#
plt.show()
```
